```python
json_data = json.loads(json_str)
```

```python
video_url = json_data['microformat']['playerMicroformatRenderer']['embed']['iframeUrl']
```

```python
#audio_url = json_data['streamingData']['adaptiveFormats'][-2]['url']
```

```python
title = json_data['microformat']['playerMicroformatRenderer']['title']['simpleText']
```

```python
title = title.replace(' ', '')
```

```python
title = re.sub(r'[\/:|?*"<>]', '', title)
print(video_url)
print(title)
```

```python
video = requests.get(video_url)
```

```python
with open(f'{title}.mp4', mode='wb') as f:
    for video_chunk in video.iter_content(1024*1024*2):
        f.write(video_chunk)
```
