```python
def granular_synthesis(audio_file, grain_size=1024, overlap=0.5, shift_range=0.1):
    # Load the audio file
    rate, audio = wavfile.read(audio_file)
    
    # Calculate the overlap size
    overlap_size = int(grain_size * overlap)
    
    # Calculate the number of grains
    num_grains = int(np.ceil((len(audio) - overlap_size) / (grain_size - overlap_size)))
    
    # Calculate the start positions of each grain
    starts = np.arange(0, len(audio) - grain_size + 1, grain_size - overlap_size)
    
    # Initialize the output signal
    output = np.zeros(len(audio))
    
    # Create the grains one at a time and add them to the output
    for i, start in enumerate(starts):
        end = start + grain_size
        
        # Initialize the grain
        grain = np.zeros(grain_size)
        
        # Extract the grain from the audio
        grain[:grain_size - overlap_size] = audio[start:end - overlap_size]
        grain[grain_size - overlap_size:] = audio[end - overlap_size:end]
        
        # Apply a window function to the grain (e.g., Hann window)
        window = np.hanning(grain_size)
        np.multiply(grain, window, out=grain)
        
        # Apply a random shift to the grain
        shift_amount = np.random.uniform(-shift_range, shift_range)
        shift_sample = int(np.round(shift_amount * grain_size))
        shifted_grain = np.zeros_like(grain)
        if shift_sample >= 0:
            shifted_grain[shift_sample:] = grain[:grain_size - shift_sample]
        else:
            shifted_grain[:grain_size + shift_sample] = grain[-shift_sample:]
        
        # Add the shifted grain to the output signal
        output[start:end - overlap_size] += shifted_grain[:grain_size - overlap_size]
        output[end - overlap_size:end] += shifted_grain[grain_size - overlap_size:]
    
    # Normalize the output signal
    output /= np.max(np.abs(output))
    
    return rate, output
```

```python
def granular_synthesis(audio_file, grain_size=1024, overlap=0.5, shift_range=0.1):
    # Load the audio file
    rate, audio = wavfile.read(audio_file)
    
    # Calculate the overlap size
    overlap_size = int(grain_size * overlap)
    
    # Calculate the number of grains
    num_grains = int(np.ceil((len(audio) - overlap_size) / (grain_size - overlap_size)))
    
    # Initialize the grains array
    grains = np.zeros((num_grains, grain_size))
    
    # Calculate the start positions of each grain
    starts = np.arange(0, len(audio) - grain_size + 1, grain_size - overlap_size)
    
    # Extract the grains from the audio
    for i, start in enumerate(starts):
        end = start + grain_size
        grains[i, :grain_size - overlap_size] = audio[start:end - overlap_size]
        grains[i, grain_size - overlap_size:] = audio[end - overlap_size:end]
    
    # Apply a window function to the grains (e.g., Hann window)
    window = np.hanning(grain_size)
    grains *= window.reshape((1, -1))
    
    # Apply a random shift to each grain
    shift_amounts = np.random.uniform(-shift_range, shift_range, num_grains)
    shift_samples = np.round(shift_amounts * grain_size).astype(int)
    
    # Shift the grains
    shifted_grains = np.zeros_like(grains)
    for i, shift in enumerate(shift_samples):
        shifted_grains[i] = np.roll(grains[i], shift)
    
    # Shuffle the shifted grains to generate random talking sounds
    np.random.shuffle(shifted_grains)
    
    # Calculate the output signal size
    output_size = (num_grains - 1) * (grain_size - overlap_size) + grain_size
    
    # Initialize the output signal
    output = np.zeros(output_size)
    
    # Add the shifted grains to the output signal
    for i, start in enumerate(starts):
        end = start + grain_size
        output[start:end] += shifted_grains[i]
    
    # Normalize the output signal
    output /= np.max(np.abs(output))
    
    return rate, output

# Example usage
rate, output = granular_synthesis("C:/Users/ASUS/Speech.wav", grain_size=2048, overlap=0.5, shift_range=0.5)

# Extend the new sound to 10 seconds
new_output = np.tile(output, 10)
wavfile.write("C:/Users/ASUS/112Random_Talking_0.5-0.5-4096.wav", rate, new_output)
```
