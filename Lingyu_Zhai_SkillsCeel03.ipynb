```python
def PM(A):
  n = len(A)
  m = n/2
  A11 = [[0 for i in range(m)]for j in range(m)]
  A12 = [[0 for i in range(m)]for j in range(m)]
  A21 = [[0 for i in range(m)]for j in range(m)]
  A22 = [[0 for i in range(m)]for j in range(m)]
  for x in range(0,m):
    for y in range(0,m):
      A11[x][y] = A[x][y]
      A12[x][y] = A[x][y+m]
      A21[x][y] = A[x+m][y]
      A22[x][y] = A[x+m][y+m]
  return (A11,A12,A21,A22)
```

```python
def MM(A11,A12,A21,A22):
  m = len(A11)
  n = 2*m
  A = [0 for i in range(n)]
  for x in range(0,n):
    for y in range(0,n):
      if x <= (m-1) and y <= (m-1):
        A[x][y] = A11[x][y]
      elif x <= (m-1) and y > (m-1):
        A[x][y] = A12[x][y-m]
      elif x > (m-1) and y<= (m-1):
        A[x][y] = A21[x-m][y]
      else:
        A[x][y] = A22[x-m][y-m]
  return A
```

```python
def AM(A,B):
  n = len(A)
  C = [[0 for i in range(n)]for j in range(n)]
  for i in range (0,n):
    for j in range (0,n):
      C[i][j] = A[i][j]+B[i][j]
  return C  
```
